```python
import os
from PIL import Image

def process_images_and_labels(base_path):
    # Paths to images and labels folders
    images_folder = base_path  # Folder where images are saved
    labels_folder = os.path.join(base_path, 'labels')  # Folder where label files are saved

    # Check if images and labels folders exist
    if not os.path.exists(images_folder) or not os.path.exists(labels_folder):
        raise FileNotFoundError("Images or labels folder not found!")

    # List all label files
    label_files = [f for f in os.listdir(labels_folder) if f.endswith('.txt')]

    for label_file in label_files:
        # Load the label file
        label_file_path = os.path.join(labels_folder, label_file)
        with open(label_file_path, 'r') as f:
            label_data = f.readlines()

        # Find the corresponding image file
        image_file = label_file.replace('.txt', '.png')
        image_file_path = os.path.join(images_folder, image_file)

        if not os.path.exists(image_file_path):
            # Try with '.jpg' extension if '.png' not found
            image_file = label_file.replace('.txt', '.jpg')
            image_file_path = os.path.join(images_folder, image_file)

            if not os.path.exists(image_file_path):
                print(f"No corresponding image found for: {label_file}")
                continue

        # Open the image to get its dimensions
        image = Image.open(image_file_path)
        image_width, image_height = image.size

        # Process each bounding box in the label file
        for idx, line in enumerate(label_data):
            elements = line.strip().split()
            if len(elements) < 5:
                print(f"Invalid label format in file: {label_file}")
                continue
            class_id = int(elements[0])  # Bounding box class
            x_center_norm = float(elements[1])  # Normalized x center coordinate
            y_center_norm = float(elements[2])  # Normalized y center coordinate
            width_norm = float(elements[3])     # Normalized bounding box width
            height_norm = float(elements[4])    # Normalized bounding box height

            # Convert normalized coordinates to pixels
            x_center = x_center_norm * image_width
            y_center = y_center_norm * image_height
            width = width_norm * image_width
            height = height_norm * image_height

            # Calculate the coordinates of the top-left and bottom-right corners
            x_min = int(x_center - (width / 2))
            y_min = int(y_center - (height / 2))
            x_max = int(x_center + (width / 2))
            y_max = int(y_center + (height / 2))

            # Ensure coordinates are within image boundaries
            x_min = max(0, x_min)
            y_min = max(0, y_min)
            x_max = min(image_width, x_max)
            y_max = min(image_height, y_max)

            # Crop the image based on the calculated coordinates
            cropped_image = image.crop((x_min, y_min, x_max, y_max))

            # Resize the cropped image to 128x128
            resized_image = cropped_image.resize((128, 128))

            # Create the name of the cropped image file and save it
            cropped_image_name = f'cropped_{os.path.splitext(image_file)[0]}_{idx}_class_{class_id}.png'
            cropped_image_path = os.path.join(images_folder, cropped_image_name)
            resized_image.save(cropped_image_path)

            print(f"Cropped and resized image saved at: {cropped_image_path}")

# Example usage:
process_images_and_labels('/content/drive/MyDrive/datasettt/inference_results/predict')
```
